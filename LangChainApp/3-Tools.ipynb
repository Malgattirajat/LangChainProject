### Trying out differnt tools and their Integration

```python
import os
from dotenv import load_dotenv
from langchain.tools import tool
from langchain_groq import ChatGroq
load_dotenv()

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

model_groq = ChatGroq(model="llama-3.1-8b-instant")

@tool
def findWeather(city:str) -> str:
    """Get a weather of the city"""
    return f"It is hot in {city} today"

model_with_tools = model_groq.bind_tools([findWeather])
response = model_with_tools.invoke("how is weather like in Hyderabad!")


###Tool Execution Loop
messages = [ {"role": "system", "content": "You are a helpful assistant. Use the findWeather tool if needed. Once you have the weather, summarize it directly to the user without doing any further research."},
    {"role": "user", "content": "what the weather like in Hyderabad"}]
ai_msg = model_with_tools.invoke(messages)
messages.append(ai_msg)

for tool_call in ai_msg.tool_calls:
    tool_result = findWeather.invoke(tool_call)
    messages.append(tool_result)

final_response = model_with_tools.invoke(messages)
final_response
```
