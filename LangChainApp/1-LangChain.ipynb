```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.agents import create_agent

def get_weather(city:str) -> str:
    """Get the weather for a city"""
    return f"the weather is cold in {city}"

agent = create_agent(
   model = "google_genai:gemini-2.5-flash",
    tools = [get_weather],
    system_prompt = "Hello World!"

)
```

```python
from langchain_core import messages
###run the code
res = agent.invoke({"messages":[{"role":"user","content":"Hyderabad weather now?" }]})
print(res["messages"][-1].content)
```

```python
### LETS GET ALL MODELS IN GROQ, GEMINI, AND OPENAI SO THAT IT IS EASY FOR US TO USE MODEL NAMES

from openai import OpenAI
from groq import Groq
from google import genai
import os

client = genai.Client(api_key=os.getenv("GOOGLE_API_KEY"))
print("=====================GEMINI MODELS=====================")
for m in client.models.list():
    print(m.name)


print("=====================OPENAI MODELS=====================")
client1 = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
for m in client1.models.list():
    print(m.id)


print("=====================GROQ MODELS=====================")
client2 = Groq(api_key=os.getenv("GROQ_API_KEY"))
modelObj = client2.models.list().data
for i in modelObj:
    print(i.id)
```
