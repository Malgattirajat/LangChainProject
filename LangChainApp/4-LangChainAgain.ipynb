### Learning LangChain

```python
import os
from dotenv import load_dotenv
from langchain.tools import tool
from langchain_groq import ChatGroq
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model_groq = ChatGroq(model="llama-3.1-8b-instant")
```

```python
from email import message
from langchain_core.messages import AIMessage, SystemMessage, HumanMessage
### Working code

student_details_1 = "A boy Raj is a class 10 student who studies in 'Delhi Public School' and he is a topper in his class of 40 people. His roll number is 27 and he was born in Hyderabad!"
student_details_2 = "A class 12th boy Ramesh is having a roll number 28 in class 10 in Delhi Public School and he has 40 people in his class. His place of birth is Bengaluru"
studetn_details_array = [student_details_1, student_details_2]

prompt = f"""From the details I give you about a boy, you must return me a json of below details- 
"name": 
"gender":
"class":
"place":
"classStrength:
"rollNo":
, and here you go for details - {student_details_1}"""

messages = [SystemMessage("You pay huge attention to details"),
HumanMessage(prompt)]

response = model_groq.invoke(messages)
print("resposne 1 -> ")
print(response)

student_function_call = [
    {
    
        "name": "get_student_info",
        "description": "Get students info from the given details",
        "parameters": {
            "type": "object",
            "properties": {
                "name": {
                    "type": "string",
                    "description": "name of the student",
                },
                "gender": {
                    "type": "string",
                    "description": "gender of the student",
                },
                "class": {
                    "type": "string",
                    "description": "class of the student",
                },
                "placeofbirth": {
                    "type": "string",
                    "description": "place of birth of the student",
                },
                "classStrength": {
                    "type": "string",
                    "description": "number of students in his class",
                },
                "rollNo": {
                    "type": "string",
                    "description": "roll number of the student",
                }
            }
        }
    }
]

model_with_tools = model_groq.bind_tools(student_function_call)

for student in studetn_details_array:
    json_response = model_with_tools.invoke([SystemMessage("You pay huge attention to details"),
HumanMessage(content=student)])
    print(json_response.tool_calls[0])
    
```
