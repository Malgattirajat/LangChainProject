### MODEL INTEGRATION WITH GEMINI, GPT, AND GROQ MODELS





```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(model='google_genai:gemini-2.5-flash')
response = model.invoke("why is india hot")
print(response.content)

groq_model = init_chat_model(model='groq:llama-3.1-8b-instant')
response1 = groq_model.invoke("how are himalayas formed")
print(response1.content)

### Couldn't use OpenAI as I am not able to find a free model of it
```

```python
### Using Google and Groq Chat
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_groq import ChatGroq
from langchain_core.messages import AIMessage, SystemMessage, HumanMessage

messages = [SystemMessage("You are a mother of two children of age 1 and 9"),
HumanMessage("both children are hungry, whom will you breatfeed?")]
    



model_google = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
response_google = model_google.invoke(messages)
print("google response -> "+response_google.content)


model_groq = ChatGroq(model="llama-3.1-8b-instant")
response_groq = model_groq.invoke(messages)
print("groq response-> "+response_groq.content)
```

```python
### Streaming
for c in model_groq.stream("What is reactive programming"):
    print(c.content+"===================================")

### Batching
responses = model_groq.batch([
    "explain yourself in 10 words",
    "explain existence in 10 words",
    "does christianity believe in dwaita or adwaita, explain in 50 words"
],
config = {
    "max_concurrency" : 2
})

for c in responses:
    print(c.content)
```
